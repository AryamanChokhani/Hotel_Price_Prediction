```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('./mumbai_hotel_prices.csv')
```

```python
dataset
```

```python
month_mapping = {'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June': 6, 'July': 7, 'August': 8, 'September': 9, 'October': 10, 'November': 11, 'December': 12}
dataset['month'] = dataset['month'].map(month_mapping)
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()

# the below line creates a multiple linear regression model and trains the model on the training set
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2) # used to display values upto 2 decimal places. It is easier to visualize that way
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)),1))
```

```python
user_input = [[month_mapping.get("July"), 5]]
regressor.predict(user_input)
```

```python
import pickle
pickle_out = open('model.pkl', 'wb')
pickle.dump(regressor, pickle_out)
pickle_out.close()
```
